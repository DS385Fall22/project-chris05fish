We need a basic understanding of git inorder to understand pull requests. Git helps development teams have version control, this is when you have multiple people changing/updating the same source code. Git helps with this and can let multiple people have their own version of source code. When it comes time to merge your changes with others and sumbit your code contributions to an on going project where you are not the sole developer, this is called a pull request.

Step 3: cd into this cloned repo and create a new branch. You can do this by entering the fowllowing command once in the repo:

```python
git checkout -b new_branch_name
```

Step 4: Now we will make some changes to any of the files in this repo and then push our changes on this new branch to github by doing the following commands:

```python
git add .
git commit -m "Your messages"
git push --set-upstream origin your_branch_name
```

Step 5: Compare and pull request by clicking the green button after going back to the repo on GitHub:

```python
display.Image("./Pull.png")
```

Step 6: Once you clicked this button you can leave a description of the changes you made and click the green button Create pull request.

Now you have created a pull request you can view by clicking the tab Pull requests. This shows details on all pull requests such as the files changed and descriptions. You can also close the pull request if something went wrong and you no longer want to attempt to merge your code.

This allows the owner of the respitory to review the pull request and then decide whether or not to accept it. After they accept your pull request, your code will be merged and displayed into the default master branch. 

Overall pull requests help single out what changes you made to the code. It also allows others to give feedback on these changes. It also creates an opportunity to make sure that the code being merged into the final source code wont cause any problems.
